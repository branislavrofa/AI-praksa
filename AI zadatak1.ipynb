```python
import pandas as pd
```

```python
df_name = pd.read_csv('https://datasets.imdbws.com/name.basics.tsv.gz', sep='\t', na_values='\\N', usecols=['nconst', 'primaryName'])
df_principals = pd.read_csv('https://datasets.imdbws.com/title.principals.tsv.gz', sep='\t', na_values='\\N', usecols=['tconst', 'nconst'])
df_movie = pd.read_csv('https://datasets.imdbws.com/title.basics.tsv.gz', sep='\t', na_values='\\N', usecols=['tconst', 'titleType', 'primaryTitle'], low_memory=False)
df_rating = pd.read_csv('https://datasets.imdbws.com/title.ratings.tsv.gz', sep='\t', na_values='\\N', usecols=['tconst', 'averageRating'])
```

```python
movie_rating = pd.merge(df_movie,df_rating,on='tconst')
```

```python
rated_movies = movie_rating.sort_values(by='averageRating',ascending=False)
```

```python
actor_name = df_name[df_name['primaryName']=='Keanu Reeves']
actor_id = actor_name['nconst'].values[0]
```

```python
rated_movies = pd.merge(rated_movies,df_principals[df_principals['nconst']== actor_id], on='tconst')
```

```python
rated_movies = rated_movies[rated_movies['titleType'] == 'movie']
```

```python
best20 = int(len(rated_movies)*0.2)
```
