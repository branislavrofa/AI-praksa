```python
import pandas as pd

df = pd.read_csv("hf://datasets/Madnesss/manga-query/data-manga.csv")
```

```python
df.head()
```

```python
space = ['space','vampire','cosmic','alien','galaxy','sci-fi']
vampire = ['vampire','blood','dracula','undead','vampires']

target = space + vampire 
```

```python
def similarity(row):
    a = str(row['description']).lower() + " " + str(row['tags']).lower()
    pom = 0
    for word in target:
        if word in a:
            pom +=1
    return pom
```

```python
df['similarity'] = df.apply(similarity,axis=1)
```

```python
top50 = df.sort_values(by ='similarity',ascending= False).head(50)
```
