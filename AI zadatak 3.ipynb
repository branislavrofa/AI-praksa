```python
import pandas as pd

df = pd.read_csv("hf://datasets/Ammok/walmart_sales_prediction/Walmart.csv")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as mse
```

```python
df['Date'] = pd.to_datetime(df['Date'] , format = '%d-%m-%Y')
df['mesec'] = df['Date'].dt.month
```

```python
srednja = df.groupby('Store')['Weekly_Sales'].mean()
df['srednja_zarada'] = df['Store'].map(srednja)
```

```python
train = df[(df['Date'] >= '2010-02-05' ) & (df['Date']<= '2012-06-15')]
test = df[(df['Date'] < '2010-02-05') | (df['Date'] > '2012-06-15')]
```

```python
X = ['mesec','Holiday_Flag','Temperature','Fuel_Price','CPI','Unemployment','srednja_zarada']
```

```python
X_train = train[X]
y_train = train['Weekly_Sales']
X_test = test[X]
y_test = test['Weekly_Sales']
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
print(mse(y_test, y_pred))
```
